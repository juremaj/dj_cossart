```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
imaging_root_data_dir = '/media/cossart/DATA/dj_cossart/data'
subject_str = 'jm003'
session_str = '2022-05-05_b'

# all_plane = ['combined', 'plane0', 'plane1', 'plane2']
all_plane = ['plane0'] 
```

```python
for plane in all_plane:
    f = np.load(f'{imaging_root_data_dir}/{subject_str}/{session_str}/suite2p/{plane}/F.npy')
    spks = np.load(f'{imaging_root_data_dir}/{subject_str}/{session_str}/suite2p/{plane}/spks.npy')
    
    for i in range(10):
        plt.figure(figsize=(10,1), dpi=200)
        plt.plot(f[i,:])
        plt.plot(spks[i,:])
        plt.title(f'Plane: {plane}')
```

```python
from scipy.stats import zscore
```

```python
plt.figure(figsize=(20,10), dpi=300)
plt.imshow(zscore(f,1), cmap='Greys', vmin=0, vmax=3.291)
plt.xlabel('Time (stamps at acquisition rate)')
plt.ylabel('Neuron index')
```
