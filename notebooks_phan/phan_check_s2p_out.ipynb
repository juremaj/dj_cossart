```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
imaging_root_data_dir = '/media/cossart/DATA/dj_cossart/data'
subject_str = 'jm007'
session_str = '2022-06-13_b'
tseries_name = 'TSeries-05182022-1006-002'

# all_plane = ['combined', 'plane0', 'plane1', 'plane2']
all_plane = ['plane0'] 
```
